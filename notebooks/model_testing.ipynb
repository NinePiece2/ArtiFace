```python
for i in [65, 75]:
  generator = tf.keras.models.load_model(f'../checkpoints/generator_epoch_{(i)}.h5')

  ##Generate 10 images and plot with no axes
  noise = np.random.normal(0, 1, (10, 250))
  gen_imgs = generator.predict(noise)
  gen_imgs = 0.5 * gen_imgs + 0.5

  fig, axs = plt.subplots(2, 5, figsize=(15, 6))
  fig.subplots_adjust(hspace = .5, wspace=.001)
  axs = axs.ravel()
  for i in range(10):
      axs[i].imshow(gen_imgs[i])
      axs[i].axis('off')
  plt.show()
```
